```python
def msToSec(msec):
    return float(f"{msec/1000:.2f}")

df["duration_sec"] = df["duration_ms"].apply(msToSec)
```

Since it is hard to read milliseconds from the graphs, we added a new column for the durations. We have used a function named msToSec to convert millisecond values in the rows to seconds and added this new row to the dataframe.

```python
values = []
curr_data = []
sortdata = {}
features1 = ["danceability","acousticness","duration_sec","liveness","energy","loudness"]
r=3
c=2

fig, ax = plt.subplots(3,2, figsize=(18,18))
y_pos = np.arange(len(listOfGenres))
for i in range(6):
    for ind,data in enumerate(listOfGenres):
        curr_mean = data[features1[i]].mean()
        if(features1[i] == "loudness"):
            curr_mean += 20
        values.append(curr_mean)
        sortdata[curr_mean] = genreNames[ind]

    values.sort()
    for mean in values:
        curr_data.append(sortdata[mean])
    ax[i // c][i % c].barh(curr_data, values,align='center', color="green")
    ax[i // c][i % c].set_label(features1[i])
    ax[i // c][i % c].set_title(features1[i], fontsize=15)
    values = []
    curr_data = []


plt.show()
```

#### Hypothesis Test Part A



**<font color="#1DB954">Hypothesis Test:</font>** We want to test whether there are significant differences in terms of popularity and the features duration, danceability, and tempo.

**<font color="#1DB954">Null Hypothesis ( H0 ):</font>** Mean popularity for all target values are the same. 

**<font color="#1DB954">Alternative Hypothesis ( HA ):</font>** Mean popularity for all target values are different.

**<font color="#1DB954">Significance level:</font>** As most hypothesis tests assume a significance level of 0.05, we are setting it as 0.05 for our test too.

**<font color="#1DB954">Null Hypothesis ( HO ):</font>** Mean duration is the same for different popularity samples
HO: μd1=μd2

**<font color="#1DB954">Alternative Hypothesis ( HA ):</font>** Mean duration is not the same for different popularity samples
HA:  Means  μd1 and μd2  are not same.

```python
from scipy.stats import ttest_ind

sample3 =merged[merged["popularity_x"] > merged["popularity_x"].quantile(0.99995)]

ttest_ind(merged["duration_ms_x"], sample3["duration_ms_x"], equal_var=False)
```

**<font color="#1DB954">Result:</font>**
As p-value we obtained is smaller than the threshold significance level 0.05, we can conclude that mean duration is not the same. Here, we reject the null hypothesis.

```python
plt.scatter(merged["popularity_x"], merged["duration_ms_x"], color="#1DB954")

plt.xlabel("Popularity")
plt.ylabel("Duration") 

plt.grid(True)
plt.show()
```

**<font color="#1DB954">Null Hypothesis ( H₀ ):</font>** Mean danceability is the same for different popularity samples
H₀: μd1=μd2

**<font color="#1DB954">Alternative Hypothesis ( HA ):</font>** Mean danceability is not the same for different popularity samples
HA:  Means  μd1 and μd2  are not same.

```python

#sample2 = merged[merged["popularity_x"] > 70]
sample3 =merged[merged["popularity_x"] > merged["popularity_x"].quantile(0.99995)]

ttest_ind(merged["danceability_x"], sample3["danceability_x"], equal_var=False)
```

**<font color="#1DB954">Result:</font>**
As p-value we obtained is smaller than the threshold significance level 0.05, we can conclude that mean danceability is not the same. Here, we reject the null hypothesis.

```python
plt.scatter(merged["popularity_x"], merged["danceability_x"], color="#1DB954")

plt.xlabel("Popularity")
plt.ylabel("Danceability") 

plt.grid(True)
plt.show()
```

**<font color="#1DB954">Null Hypothesis ( H₀ ):</font>** Mean tempo is the same for different popularity samples
H₀: μt1=μt2

**<font color="#1DB954">Alternative Hypothesis ( HA ):</font>** Mean tempo is not the same for different popularity samples
HA:  Means  μt1 and μt2  are not same.

```python
sample3 =merged[merged["popularity_x"] > merged["popularity_x"].quantile(0.99995)]

ttest_ind(merged["tempo_x"], sample3["tempo_x"], equal_var=False)
```

**<font color="#1DB954">Result:</font>**

**<font color="#1DB954"><h2>Song Recommendation System</h2></font>**

To make song recommendations we wanted to use the KNN method however when we tested it with given recitation codes, we failed to make good predictions since we weren't using enough features. To use KNN with multiple features efficiently we could not find a good library but instead we found the following functions. These functions just return the k-nearest neighbors of a song using the given features. Later we selected the most important features and created a song recommendation function.

The following three functions are taken here: 
https://www.kaggle.com/leomauro/spotify-songs-similarity-search

```python
# K-query
def knnQuery(queryPoint, arrCharactPoints, k):
    tmp = arrCharactPoints.copy(deep=True)
    tmp['dist'] = tmp.apply(lambda x: np.linalg.norm(x-queryPoint), axis=1)
    tmp = tmp.sort_values('dist')
    return tmp.head(k).index

# Range query
def rangeQuery(queryPoint, arrCharactPoints, radius):
    tmp = arrCharactPoints.copy(deep=True)
    tmp['dist'] = tmp.apply(lambda x: np.linalg.norm(x-queryPoint), axis=1)
    tmp['radius'] = tmp.apply(lambda x: 1 if x['dist'] <= radius else 0, axis=1)
    return tmp.query('radius == 1').index

# Execute k-NN removing the 'query point'
def querySimilars(df, columns, idx, func, param):
    arr = df[columns].copy(deep=True)
    queryPoint = arr.loc[idx]
    arr = arr.drop([idx])
    response = func(queryPoint, arr, param)
    return response
```

We selected numeric columns that we found the be generally related with song genre in part-B. We omitted some not very important features but tried to keep most significant ones available for our neighbor search.  

```python
sig_cols = ['acousticness','danceability','energy','instrumentalness','liveness','valence']
```

Recomandation function takes a song list as a paramater and recomends specfied number of new songs for each song using features above. It can find recomandations for any song as long as it is in main dataset df.

```python
def recomendSong(songList,numRecommend):
    func, param = knnQuery, numRecommend
    for song in songList:
        songIndex = df[df["name"] == song].index[0]
        curr = df.loc[songIndex]
        print("Getting recommandations for:",curr["name"], "-", curr["artists"])
        response = querySimilars(df, sig_cols, songIndex, func, param)
        for idx in response:
            curr = df.loc[idx]
            print("\n", curr["name"], "-", curr["artists"])
        print("------------")
```

```python
songList = ["When You're Smiling (The Whole World Smiles With You) - Without Intro",
            "positions", 
            "Duality"]
```

```python
recomendSong(songList, 2)
```

As it can be seen above, the algorithm recommended very similar and accurate songs for the given outputs and we can say that this algorithm works fine.

For Louis Armstrong's When You're Smiling, we got recommendations as Frank Sinatra and Bill Evans' songs which are clearly accurate recommendations since their genres are also jazz and they are very similar songs.

For Ariana Grande's positions it recommended related songs with pop elements like 2Pac.
